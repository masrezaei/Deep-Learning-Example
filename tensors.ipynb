# 2.2 Data representations for neural networks

## tensors 

### Scalars (0D tensors)

```python
import numpy as np
x = np.array(12)
print(x,)
print('rank=',x.ndim)
print('shape=',x.shape)
```

### Vectors (1D tensors)

```python
x = np.array([12, 3, 6, 14])
print(x,)
print('rank=',x.ndim)
print('shape=',x.shape)
```

### Matrices (2D tensors)

```python
x = np.array([[5, 78, 2, 34, 0],
              [6, 79, 3, 35, 1],
              [7, 80, 4, 36, 2]])
print(x,)
print('rank=',x.ndim)
print('shape=',x.shape)
```

### 3D tensors and higher-dimensional tensors

```python
x = np.array([[[5, 78, 2, 34, 0],
               [6, 79, 3, 35, 1],
               [7, 80, 4, 36, 2]],
               [[5, 78, 2, 34, 0],
               [6, 79, 3, 35, 1],
               [7, 80, 4, 36, 2]],
               [[5, 78, 2, 34, 0],
               [6, 79, 3, 35, 1],
               [7, 80, 4, 36, 2]]])
print(x,)
print('rank=',x.ndim)
print('shape=',x.shape)
```
